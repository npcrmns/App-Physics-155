```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Line Graphs

```python
x = np.linspace(0, 4*np.pi, 100)
y = np.sin(2*x)
z = np.cos(1.5*x)
```

```python
plt.plot(x, y, color="black", label="Signal A")
plt.plot(x, z, color="red", label="Signal B")
plt.grid()
plt.xlabel(r"Time $t$") # r treats the string as a raw string
plt.ylabel(r"Output $V_{out}$")
plt.legend()

# use DPI of 150 so it does not look pixelated on your papers
plt.savefig("signalgen.png", dpi=150)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10,4))

ax[0].plot(x, y, color="black")
ax[1].plot(x, y, color="red")

for a in ax:
    a.set_xlabel(r"Time $\tau$")
    a.set_ylabel(r"Output $V_{out}$")
    a.grid()
    
fig.tight_layout() # solves the issue of overlaping labels due to small size
plt.savefig("fancyfig.png", dpi=150)
```

```python
x = np.linspace(0, 4*np.pi, 100)
y = np.sin(2*x)
z = np.cos(1.5*x)

fig, ax = plt.subplots()

ax.plot(x, y, color='black', label="Signal A", ls="-.")
ax.scatter(x, z, color='red', label="Signal B")

ax.set_xlabel(r"Time $t$") 
ax.set_ylabel(r"Output $V_{out}$")

ax.grid()
ax.legend()

fig.tight_layout()
```

```python
# Always make sure x and y have same dimension when plotting
print(len(x), len(y))
```
